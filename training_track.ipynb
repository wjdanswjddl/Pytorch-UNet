```python
import pandas as pd
import matplotlib.pyplot as plt

# Define file paths
loss_file_path = '/home/abhat/wirecell_sbnd/Pytorch-UNet/checkpoints/UNet/tpc1_bothplanes_with_prolongedtrks/loss.csv'
dice_file_path = '/home/abhat/wirecell_sbnd/Pytorch-UNet/checkpoints/UNet/tpc1_bothplanes_with_prolongedtrks/dice.csv'

# Read CSV files
loss_data = pd.read_csv(loss_file_path, header=None, names=['epoch', 'train_loss', 'val_loss'])
dice_data = pd.read_csv(dice_file_path, header=None, names=['epoch', 'val_dice'])

# # Remove duplicate entries by keeping the last occurrence for each epoch
# loss_data = loss_data.drop_duplicates(subset=['epoch'], keep='last')
# dice_data = dice_data.drop_duplicates(subset=['epoch'], keep='last')

# Merge data on epoch
data = pd.merge(loss_data, dice_data, on='epoch')

# Plot training and validation loss
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(data['epoch'], data['train_loss'], label='Training Loss', color='blue', marker='o')
plt.plot(data['epoch'], data['val_loss'], label='Validation Loss', color='orange', marker='o')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.grid()
plt.legend()

# Plot validation dice coefficient
plt.subplot(1, 2, 2)
plt.plot(data['epoch'], data['val_dice'], label='Validation Dice', color='green', marker='o')
plt.xlabel('Epoch')
plt.ylabel('Dice Coefficient')
plt.title('Validation Dice Coefficient')
plt.grid()
plt.legend()

# Display the plots
plt.tight_layout()
plt.show()
```
